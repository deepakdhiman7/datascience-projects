```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import string
from sklearn.feature_extraction.text import CountVectorizer
```

```python
f=open(r'C:\Users\user\Desktop\DataScience\UCSenDiego_ML\4_LinearRegression\sentiment-logistic-regression\sentiment-logistic-regression\sentiment_labelled_sentences\full_set.txt', 'r')
content = f.readlines()
print(content)
```

```python
content = [x.strip() for x in content]
print(content[0:3])
```

```python
sentences=[x.split('\t')[0] for x in content]
labels=[x.split('\t')[1] for x in content]
```

```python
y=np.array(labels, dtype='int8')
y=(y*2)-1
print(y[0:3])
```

```python
digits=[str(i) for i in range(0,10)]
digit_less=[]
for i in sentences:
    for j in digits:
        i=i.replace(j,'')
    digit_less.append(i)
digit_less
```

```python
punc_less=[]
for i in digit_less:
    for j in list(string.punctuation):
        i=i.replace(j,'')
    punc_less.append(i)
punc_less
    
```

```python
sentence_lower=[i.lower() for i in punc_less]
sentence_lower
```

```python
sentences_splt=[i.split() for i in sentence_lower]
print(sentences_splt[0:3])
```

```python
stop_set = ['the', 'a', 'an', 'i', 'he', 'she', 'they', 'to', 'of', 'it', 'from']

sent_processed=[]
for a in sentences_splt:
    b=[s for s in a if s not in stop_set]
    sent_processed.append(" ".join(b))
sent_processed[0:10]
```

```python
vectorizer=CountVectorizer(analyzer='word',tokenizer=None, preprocessor=None,stop_words=None, max_features=4500)
# print(vectorizer)
data_features=vectorizer.fit_transform(sent_processed)
vectorizer.stop_words
```

```python
data_mat=data_features.toarray()
p_mat=data_features.toarray()
p_mat.shape
data_mat=np.ones((p_mat.shape[0],p_mat.shape[1]+1))
data_mat.shape
```

```python
data_mat[:,:-1]=p_mat
data_mat
```

```python
np.random.seed(0)
test_inds=np.append(np.random.choice(np.where(y==-1)[0], 250, replace=False), np.random.choice(np.where(y==1)[0], 250, replace=False))

train_inds=list(set(range(len(labels)))-set(test_inds))

train_data=data_mat[train_inds,]
train_labels=y[train_inds]

test_data=data_mat[test_inds,]
test_labels=y[test_inds]

print("train data: ", train_data.shape)
print("test data: ", test_data.shape)
```

```python
np.random.seed(0)
test_inds= np.append(np.random.choice((np.where(y==-1)[0]), 250, replace=False),np.random.choice((np.where(y==1)[0]), 250, replace=False))
# print(test_inds)
train_inds=list(set(range(len(labels)))-set(test_inds))
train_data = data_mat[train_inds,]
train_labels = y[train_inds]

test_data = data_mat[test_inds,]
test_labels = y[test_inds]

print("train data: ", train_data.shape)
print("test data: ", test_data.shape)
```

```python
from sklearn.linear_model import SGDClassifier

clf=SGDClassifier(loss='log', penalty='none')
clf.fit(train_data,train_labels)
w=clf.coef_
b=clf.intercept_

preds_train=clf.predict(train_data)
preds_test=clf.predict(test_data)

errs_train = np.sum((preds_train > 0.0) != (train_labels > 0.0))
errs_test = np.sum((preds_test > 0.0) != (test_labels > 0.0))

print('Training error:',float(errs_train)/len(preds_train))
print('Test error',float(errs_test)/len(preds_test))
```
